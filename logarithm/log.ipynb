## Logarithm
- It is basically inverse of exponent.
- Exponent is power of something. For example a^5 5 is exponent of a.
- The popular log is log to the base 10.
- log 10 is always 1, we have to remember that.

## How Log can be used in data analysis process?
- to transform skewed data to approximately conform to normality.
- Log transform in machine learning. Log convert all the datapoints in a comparable values. so that the model will not biased and give accurate results. 

```python
# importing libraries

import pandas as pd
import matplotlib.pyplot as plt

```

```python
# list of list
my_list = [['Tesla', 31],['Chelsea', 16],['Prime Bank', 6.0],['Daraz', 20],['Standard chartered', 5.5],['Amazon', 355],['Uber', 11]]
column = ['Company', 'Revenue']

df = pd.DataFrame(my_list, columns=column)
df.head()
```

```python
df.plot(x='Company', y='Revenue', kind='bar')
```

- It is hard to see datapoints in standard chartered and prime bank. We cannot say which has high value.
- Logarithm solves this kind of problem

```python
df.plot(x='Company', y='Revenue', kind='bar', logy=True)
```

- Now we can see bar are little more comparable than the above bar. Here we can atleast say prime bank has high value than standard chartered. 

## Log Normal Distribution
- If you get a normal distribution by applying a log function to a dataset then dataset is log normally distributed.


```python
# list of dictionary
income = [{'Income': 15000, 'Count': 10000},
          {'Income': 25000, 'Count': 8000},
          {'Income': 100000, 'Count': 7500},
          {'Income': 250000, 'Count': 6000},
          {'Income': 12100, 'Count': 6500},
          {'Income': 21000, 'Count': 5000},
          {'Income': 55000, 'Count': 4500},
          {'Income': 300000, 'Count': 4800},
          {'Income': 63000, 'Count': 3500},
          {'Income': 50500, 'Count': 2528},
          {'Income': 150000, 'Count': 1100},
          {'Income': 2500, 'Count': 900},
          {'Income': 85000, 'Count': 500},
          {'Income': 10000000, 'Count': 200}]

data = pd.DataFrame(income)
```

```python
data.head()
```

```python
import seaborn as sns
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
g = sns.barplot(x='Income',y='Count',data=data)
g.set_xticklabels(g.get_xticklabels(), 
                          rotation=45, 
                          horizontalalignment='right');
```

The bar is right skewed

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
g = sns.barplot(x='Income',y='Count',data=data)
g.set_xticklabels(g.get_xticklabels(), 
                          rotation=45, 
                          horizontalalignment='right');
g.set(xscale="log");
```

With the help of log normal distribution char looks more normarl distributed.
