```python
import pandas as pd
```

```python
dataframe = pd.read_csv('AB_NYC_2019.csv')
dataframe.head()
```

```python
# statistics of the dataframe
dataframe.price.describe()
```

```python
print(f'75 percentile is {dataframe.price.quantile(0.75)}')
print(f'100 percentile is {dataframe.price.quantile(1)}')
```

There is a huge different in 75 and 100 percentile

We can say that there is an outliers in the dataframe

we have to handle it

```python
# We use minimum and maximum threshold because it helps to remove outlier
min_threshold, max_threshold = dataframe.price.quantile([0.01, 0.999])
print(f'Minimum Threshole is {min_threshold}')
print(f'Maximum Threshole is {max_threshold}')
```

Minimum returns 30

maximum returns 3000

In this way we can compare our dataframe with minimum and maximum threshold i.e between 30 and 3000 and stay away from outliers

```python
# Let's see price below min_threshold
dataframe[dataframe['price']<min_threshold].head()
```

```python
no_outlier_df = dataframe[(dataframe.price < max_threshold) & (dataframe.price > min_threshold)]
no_outlier_df.head()
```

```python
# lets see the statistical status of newly formed dataframe

no_outlier_df.price.describe()
```
